```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read in the climate dataset
climate_df = pd.read_csv('../data/chi_ny.csv')
climate_df.head()
```

```python
climate_df.shape
```

```python
climate_df[climate_df.duplicated()]
```

```python
climate_df.dtypes
```

```python
ord_climate = climate_df[climate_df['NAME'].str.contains('OHARE')]
ord_climate.head()
```

```python
ord_climate.shape
```

```python
lga_climate = climate_df[climate_df['NAME'].str.contains('LAGUARDIA')]
lga_climate.head()
```

```python
lga_climate.shape
```

```python
# check for null values in each column
ord_climate.isnull().sum()
```

```python
# check for null values in each column
lga_climate.isnull().sum()
```

```python
# drop columns that contain null values for all or a majority of days
drop_columns = ['PGTM', 'PSUN', 'TOBS', 'TSUN', 'WDMV', 'WESD', 'WESF', 'WT10']
last_10_col = climate_df.columns[-10:].tolist()
drop_columns += last_10_col
ord_climate.drop(columns=drop_columns, inplace=True)
lga_climate.drop(columns=drop_columns, inplace=True)
```

```python
ord_climate.reset_index(drop=True, inplace=True)
ord_climate.head()
```

```python
lga_climate.reset_index(drop=True, inplace=True)
lga_climate.head()
```

```python
ord_climate.describe().T
```

```python
lga_climate.describe().T
```

The ranges of values of each feature looks correct, and therefore we will assume there are no extreme outliers that will skew the results.

```python
ord_climate.isnull().sum()
```

Two columns still have one null value.

```python
# locate the rows with null values
ord_climate[ord_climate.isnull().any(axis=1)]
```

```python
ord_climate[ord_climate['DATE'].str.contains('2015-06-1')]
```

```python
# replace the null values with the averages from three days before and after 6/16
ord_climate['WDF5'].loc[166] = ord_climate[ord_climate['DATE'].str.contains('2015-06-1')][3:]['WDF5'].mean()
ord_climate['WSF5'].loc[166] = ord_climate[ord_climate['DATE'].str.contains('2015-06-1')][3:]['WSF5'].mean()
```

```python
ord_climate.loc[166]
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(ord_climate[ord_climate.columns[3:]].corr(), cmap='coolwarm', annot=True);
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(lga_climate[lga_climate.columns[3:]].corr(), cmap='coolwarm', annot=True);
```

There are strong positive correlations found between the min, max, and average temperatures, which is to be expected. Similarily, the features associated to wind speed (AWND, WSF2, WSF5) and wind direction (WDF2, WDF5) are highly correlated as well. There seems to be a mild negative correlation between temperature features and snow depth.

```python
ord_climate[ord_climate.columns[1:16]].hist(figsize=(20,20))
plt.show;
```

```python
lga_climate[lga_climate.columns[1:16]].hist(figsize=(20,20))
plt.show;
```

The average wind speed tends to be higher in Chicago compared to New York. And the distributions of the directions of the fastest 2 minute and 5 second wind speed varies between Chicago and New York. In New York it seems the directions of wind changes drastically. Most of the other climate attributes have similar distributions between the two cities.

```python
# save cleaned data into CSV files
ord_climate.to_csv('../data/ord_climate_by_day.csv', index=False)
lga_climate.to_csv('../data/lga_climate_by_day.csv', index=False)
```
